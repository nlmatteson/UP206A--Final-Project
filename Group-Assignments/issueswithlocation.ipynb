# Prearing Datasets for plotting

```python
import pandas as pd

# for spatial data
import geopandas as gpd

# for plotting
import matplotlib.pyplot as plt
# for basemaps
import contextily as ctx

import folium
from folium.plugins import MarkerCluster
import pandas as pd

import plotly.express as px
```

```python
ethindex=gpd.read_file("ethnicityindex.geojson")
stations=gpd.read_file("stations.geojson")
stationarea=gpd.read_file("stationarea.geojson")
```

```python
tod=stations[stations['TOD']=='Yes']
ntod=stations[stations['TOD']=='No']
```

```python
stations.crs
```

```python
stationarea_wm=stationarea.to_crs(epsg=3857)
stations_wm=stations.to_crs(epsg=3857)
```

```python
fig, ax = plt.subplots(figsize=(12,12))

# block groups
stationarea.plot(ax=ax)


ctx.add_basemap(ax,source=ctx.providers.CartoDB.Positron, zoom=10, crs=4326)
```

```python
fig, ax = plt.subplots(figsize=(12,12))

# block groups
stationarea_wm.plot(ax=ax)

tod.plot(ax=ax,
         color = 'black',
         marker = "D",
         markersize = 15,
        label = 'Station')

ax.axis('off')
ax.set_title('Change in Block Groups who spend >30% of their income on rent',fontsize=15,pad=18)

# add a basemap
ctx.add_basemap(ax,source=ctx.providers.CartoDB.Positron, zoom=10)
```
