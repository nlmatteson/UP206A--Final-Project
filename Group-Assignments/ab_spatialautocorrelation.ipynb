# Spatial autocorrelaiton
We perform spatial autocorrelation to create lagged value for the indexes. This will help us better interpret our data. 

```python
## to read and wrangle data
import pandas as pd

# to create spatial data
import geopandas as gpd

# for basemaps
import contextily as ctx

# For spatial statistics
import esda
from esda.moran import Moran, Moran_Local

import splot
from splot.esda import moran_scatterplot, plot_moran, lisa_cluster,plot_moran_simulation

import libpysal as lps

# Graphics
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
gdf=gpd.read_file("ethnicityindex.geojson")
```

We create a new variable that measures the change in diversity index

```python
gdf_bg=gdf
gdf_bg['index_change']=((gdf_bg['iqv_2019']-gdf_bg['iqv_2013'])/gdf_bg['iqv_2013'])*100
```

Creating spatial lags

```python
wq=lps.weights.KNN.from_dataframe(gdf,k=8)
wq.transform = 'r'
gdf_bg['iqv_2019_lag'] = lps.weights.lag_spatial(wq, gdf_bg['iqv_2019'])
# create a new column for the spatial lag
gdf_bg['iqv_2013_lag'] = lps.weights.lag_spatial(wq, gdf_bg['iqv_2013'])
gdf_bg['index_change_lag'] = lps.weights.lag_spatial(wq, gdf_bg['index_change'])
```

plotting non-lagged and lagged variables

```python
# create the 1x2 subplots
# set sharex and sharey to true to scale each map equally
fig, axs = plt.subplots(1, 2, figsize=(26, 50), sharex=True, sharey=True)

# name each subplot
ax1, ax2 = axs

# percent hispanic map on the left
gdf.plot(ax=ax1,
         figsize=(15,15),
         column='iqv_2019',
         legend=False,
         alpha=0.8,
         cmap='YlOrRd',
         scheme='quantiles',
        edgecolor='white',
        linewidth=0.2,)


ax1.axis("off")
ax1.set_title("Raw figures")

# percent black map on the right
gdf_bg.plot(ax=ax2,
         column='iqv_2019_lag',
         legend=False,
         alpha=0.8,
         cmap='YlOrRd',
         scheme='quantiles',
         edgecolor='white',
         linewidth=0.2)

ax2.axis("off")
ax2.set_title("With spatial autocorrelation")
ctx.add_basemap(ax1,source=ctx.providers.CartoDB.DarkMatter)
ctx.add_basemap(ax2,source=ctx.providers.CartoDB.DarkMatter)
```
