```python
import osmnx as ox # this is for open street network analysis
import pandas as pd # this is for regular data analysis
import geopandas as gpd # this is for geospatial analysis
import contextily as ctx # this is for basemaps
import matplotlib.pyplot as plt # this is for plotting
import networkx as nx # this is for network analysis
import matplotlib.pyplot as plt
%matplotlib inline
import folium
```

```python
bart = gpd.read_file('California_Rail_Stations.geojson')
bart.head()
```

```python
columns_to_keep = ['OBJECTID',
                   'LOCATION',
                   'STATION',
                   'CODE',
                   'ADDRESS',
                   'ZIP',
                   'COMM_NETWO',
                   'STATION_TY',
                   'DIST',
                   'CO',
                   'geometry']
```

```python
bart = bart[bart.CO == 'ALA']
bart.sample(10)
```

```python
focus_stations = bart.set_index("STATION") #this is to set the station names as the index so it is easier to delete the rows I don't need
focus_stations.head()
```

```python
focus_stations.loc[['ASHBY', 'MACARTHUR', 'HAYWARD', 'SOUTH HAYWARD', 'FRUITVALE']]
```

```python
focus_stations = focus_stations.loc[['ASHBY', 'MACARTHUR', 'HAYWARD', 'SOUTH HAYWARD', 'FRUITVALE']]
focus_stations.head()
```

```python
focus_stations = focus_stations[focus_stations.COMM_NETWO == 'BART']
focus_stations.head()
```

```python
minx, miny, maxx, maxy = focus_stations.geometry.total_bounds
```

```python
longitude = (maxx-minx)/2 + minx
latitude = (maxy-miny)/2 + miny
print(longitude)
print(latitude)
```

```python
m = folium.Map(location=[latitude,longitude], zoom_start=11) 
```

```python
m
```

```python
stations=focus_stations.reset_index()
```

```python
stations.head()
```

```python
stations['lon']=stations.geometry.x
stations['lat']=stations.geometry.y
```

```python
for index, row in stations.iterrows():
    # add folium marker code
    folium.Marker([row.lat, row.lon], popup=row.STATION, tooltip=row.STATION).add_to(m)
m
```

```python
stations.to_file("stations.geojson", driver='GeoJSON')
```
