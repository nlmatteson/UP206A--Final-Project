# **Analyzing changes in Rent Burden for Census Tracts in Alameda County: 2011 - 2019**

In this notebook, I will be looking at rent burden changes in Alameda County. Ultimately, I will create a percent change analysis to determine which census tracts had the most change regarding rent burden. 

The data I collected for 2019 and 2011 were from data.census.org. The rent burden data is gross rent as a percentage of household income. The data is broken down into groups by census tracts.

2011: https://data.census.gov/cedsci/table?q=Alameda%20County,%20California&t=Renter%20Costs&g=0500000US06001%241400000&y=2011&tid=ACSDT5Y2011.B25070
2019: https://data.census.gov/cedsci/table?q=Alameda%20County,%20California&t=Renter%20Costs&g=0500000US06001%241400000&y=2019&tid=ACSDT5Y2019.B25070

They are both CSV files, so in order to produce maps with this data, I will need to merge them with a cenus tract shapefile. 
Census tract Shapefile:

I'm first going to clean up the datasets and then merge them together into one. Then I will merge the data with the shapefile so that I can produce maps with the data. I think it will be really interesting to see which census tracts have changed the most, and then compare them to the specific BART stations our project is focusing on. This will provide some insight as to whether or not the TOD projects have impacted the surrounding areas rent burdern, either in a good way or bad.
